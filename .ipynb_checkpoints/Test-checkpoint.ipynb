```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
InstaDec22Data = pd.read_csv('./Datasets/social media influencers-INSTAGRAM - -DEC 2022.csv')
InstaDec22Data=InstaDec22Data.rename(columns={'Eng. (Auth.)':'EngAuth', 'Eng. (Avg.)':'EngAvg', 'instagram name':'realname', 'name':'username'})
# YtbDec22Data = pd.read_csv('./Datasets/social media influencers-YOUTUBE - --DEC 2022.csv')
InstaDec22Data
```

```python
# YtbDec22Data.columns
# InstaDec22Data.assign(avg= (InstaDec22Data.EngAuth + InstaDec22Data.EngAvg)/(2*InstaDec22Data.followers))
data=np.empty((1000,0))
y=np.empty((1000,1))
for arr in [InstaDec22Data.EngAuth, InstaDec22Data.EngAvg, InstaDec22Data.followers]:
    x=list(zip(arr.replace(['M$', 'K$'], '', regex=True), arr.replace(['\d', '\d', '\.'], '', regex=True)))
    for index in range(0, 1000):
        if(x[index][1]=='M'):
            y[index]=float(x[index][0])*1000000
        elif(x[index][1]=='K'):
            y[index]=float(x[index][0])*1000
        # print(y[index])
    data=np.concatenate([data, y], axis=1)
# print(data.shape)
Modified_data=pd.DataFrame(data, columns=['EngAuth','EngAvg','Followers'])
Modified_data=Modified_data
Modified_data.shape
```

```python
Final_data=Modified_data.assign(Score=((Modified_data.EngAvg/Modified_data.Followers)+(Modified_data.EngAuth)))
Final_data=Final_data.drop(['EngAuth','EngAvg','Followers'], axis=1)
```

#### Bigger the better Score

```python
# print(InstaDec22Data.isna().sum())
x=InstaDec22Data['country'].sort_values().unique()
lent = len(x)
Countries=dict()
for i in range(0,lent):
    Countries[x[i]]=i
Countries
Final_data=Final_data.assign(Country=InstaDec22Data.country)
Final_data=Final_data.assign(username=InstaDec22Data.username)
Final_data.replace(Countries)
Final_data=Final_data.sort_values('Score', ascending=False).drop_duplicates('username').sort_index()
reviewmatrix = Final_data.pivot(index="username", columns="Country", values="Score").fillna(0)
reviewmatrix
```

```python
from scipy.sparse.linalg import svds
matrix = reviewmatrix.values
U, sigma, Vt = svds(matrix, k=20)
```

```python
# predicted_scores = user_factor_matrix @ singular_values @ influencer_factor_matrix.T
user_item_predicted = np.dot(np.dot(U, np.diag(sigma)), Vt)
```
