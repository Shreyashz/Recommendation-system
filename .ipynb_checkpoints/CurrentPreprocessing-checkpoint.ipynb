```python
import os
from math import ceil
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial.distance import cosine
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

InstaDec22Data = pd.read_csv('./Datasets/social media influencers-INSTAGRAM - -DEC 2022.csv')
InstaDec22Data=InstaDec22Data.rename(columns={'Eng. (Auth.)':'EngAuth', 'Eng. (Avg.)':'EngAvg', 'instagram name':'realname', 'name':'username'})
# YtbDec22Data = pd.read_csv('./Datasets/social media influencers-YOUTUBE - --DEC 2022.csv')
InstaDec22Data

for i in range(0, len(InstaDec22Data)):
    if(InstaDec22Data.country.isna().iloc[i]):
        InstaDec22Data.country.iloc[i]= 'South Korea'
        # print(InstaDec22Data.iloc[i])
InstaDec22Data.Category_1=InstaDec22Data.Category_1.fillna("Any")
InstaDec22Data.Category_2=InstaDec22Data.Category_2.fillna("Any")
# InstaDec22Data.Category_2.isna().sum()

# YtbDec22Data.columns
# InstaDec22Data.assign(avg= (InstaDec22Data.EngAuth + InstaDec22Data.EngAvg)/(2*InstaDec22Data.followers))
data=np.empty((1000,0))
y=np.empty((1000,1))
for arr in [InstaDec22Data.EngAuth, InstaDec22Data.EngAvg, InstaDec22Data.followers]:
    x=list(zip(arr.replace(['M$', 'K$'], '', regex=True), arr.replace(['\d', '\d', '\.'], '', regex=True)))
    for index in range(0, 1000):
        if(x[index][1]=='M'):
            y[index]=float(x[index][0])*1000000
        elif(x[index][1]=='K'):
            y[index]=float(x[index][0])*1000
        # print(y[index])
    data=np.concatenate([data, y], axis=1)
# print(data.shape)
Modified_data=pd.DataFrame(data, columns=['EngAuth','EngAvg','Followers']).assign(Rank=InstaDec22Data['Rank'])
Modified_data=Modified_data.assign(Country=InstaDec22Data['country'])


data=np.empty((1000,0))
y=np.empty((1000,1))
for arr in Modified_data.index:
    y[arr]=ceil(0.001*Modified_data.Followers.iloc[arr]+1000000*Modified_data.EngAuth.iloc[arr]/Modified_data.EngAvg.iloc[arr])
    # print("==>", y[arr])
data=np.concatenate([data, y], axis=1)
Final_data=pd.DataFrame(data, columns=['Est_Pay'])
Final_data=Final_data.assign(catagory=InstaDec22Data['Category_1'])
Final_data=Final_data.assign(catagoryExta=InstaDec22Data['Category_2'])
# Modified_data=pd.DataFrame(data, columns=['EngAuth','EngAvg','Followers']).assign(Rank=InstaDec22Data['Rank'])
Final_data=Final_data.assign(Country=InstaDec22Data['country'])
Final_data=Final_data.assign(influencer=InstaDec22Data['username'])
Final_data=Final_data.assign(EngAvg=Modified_data['EngAvg'])
Final_data=Final_data.assign(Followers=Modified_data['Followers'])
Final_data.to_csv('Final_data.csv', index=False)

scaler = MinMaxScaler()
training_data = Final_data
training_data[['Followers', 'EngAvg', 'Est_Pay']] = scaler.fit_transform(Final_data[['Followers', 'EngAvg', 'Est_Pay']])
def recommend_wnums(user_profile, df=training_data, top_n=10):
    user_profile_scaled = scaler.transform([[user_profile['min_followers'], user_profile['min_engagement'], user_profile['max_cost']]])
    min_followers_scaled, min_engagement_scaled, max_cost_scaled = user_profile_scaled[0]
    # print(user_profile_scaled) 
    similarities = {}
    for index, row in df.iterrows():
        influencer_vector = [row['Followers'], row['EngAvg'], row['Est_Pay']]
        user_profile_vector = [min_followers_scaled, min_engagement_scaled, max_cost_scaled]
        similarities[row['influencer']] = 1 - cosine(influencer_vector, user_profile_vector)
    # print(similarities)
    sorted_influencers = sorted(similarities.items(), key=lambda x: x[1], reverse=True)[:top_n]
    return sorted_influencers

import joblib
joblib.dump(scaler, 'scaler_model.pkl')
```
